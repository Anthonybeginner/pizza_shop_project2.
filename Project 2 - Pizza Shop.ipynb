```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://en.wikipedia.org/wiki/Pizza"
html = requests.get(URL)
URL_source_code = html.text
soup = BeautifulSoup(URL_source_code,'html.parser')
```

```python
html
```

```python
print(soup)
```

```python
soup.find_all('table')[2]
```

```python
soup.find('table', class_ = "wikitable sortable" )
```

```python
table = soup.find('table', class_ = "wikitable sortable" )
print(table)
```

```python
table.find_all('th')
```

```python
world_titles = table.find_all('th')
print (world_titles)
```

```python
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    indiv_row_data = [data.text.strip() for data in row_data]
    print(indiv_row_data)
    
    length = len(df)
    df.loc[length] = indiv_row_data
```

```python
df
```

```python
df.drop_duplicates()
```
